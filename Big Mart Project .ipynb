```python
#Dummy Variables:
sdf_sales = pd.get_dummies(sdf_sales, columns =['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type',])
sdf_sales.dtypes
```

```python
sdf_sales
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
sdf_sales['Item_Type']=le.fit_transform(sdf_sales['Item_Type'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x1 = sdf_sales[['Item_Visibility','Item_Type']]
y1 = sdf_sales['Item_Outlet_Sales'].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1, test_size = 0.33, random_state=42)
mlp = LinearRegression()
mlp.fit(x1_train,y1_train)
#y1_predict = pd.DataFrame(mlp.predict(x1_test))
#y1_test = pd.DataFrame(y1_test)
```

```python
mlp.coef_
```

```python
mlp.intercept_
```

```python
mlp.score(x1_train,y1_train)
```

```python
pred=mlp.predict(x1_test)
print('predicted result price:',pred)
print('actual price',y1_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print('error:')
print('mean absolute error:', mean_absolute_error(y1_test,pred))
print('Mean squared error:',mean_squared_error(y1_test,pred))
print('Root Means Squared Error:',np.sqrt(mean_squared_error(y1_test,pred)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y1_test,pred))
```

```python
from sklearn.linear_model import Lasso,Ridge
```
