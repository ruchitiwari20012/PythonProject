```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
x = data['Economy (GDP per Capita)'].values.reshape(-1,1)
y = data['Happiness Score'].values.reshape(-1,1)


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.33, random_state=0)

lin_reg.fit(x_train,y_train)
y_pred = lin_reg.predict(x_test)
plt.plot(x_test,y_pred)
```

```python
x1 = data[['Country', 'Region', 'Happiness Rank','Standard Error','Economy (GDP per Capita)','Family','Health (Life Expectancy)', 
            'Freedom', 'Trust (Government Corruption)','Generosity', 'Dystopia Residual']].values
y1 = data['Happiness Score'].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1, test_size = 0.33, random_state=0)
mlp = LinearRegression()
mlp.fit(x1_train,y1_train)
#y1_predict = pd.DataFrame(mlp.predict(x1_test))
#y1_test = pd.DataFrame(y1_test)
```

```python
mlp.coef_
```

```python
mlp.intercept_
```

```python
mlp.score(x1_train,y1_train)
```

```python
#predict the values 
pred=mlp.predict(x1_test)
print('predicted result price:',pred)
print('actual price',y1_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print('error:')
print('mean absolute error:', mean_absolute_error(y1_test,pred))
print('Mean squared error:',mean_squared_error(y1_test,pred))
print('Root Means Squared Error:',np.sqrt(mean_squared_error(y1_test,pred)))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y1_test,pred))
```
