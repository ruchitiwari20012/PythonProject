```python
#Now by using multiple algorithms we are calculating the best algo which suit best for our dataset
model = [DecisionTreeClassifier(),KNeighborsClassifier(),SVC()]
max_accuracy_score=0
for r_state in range(42,90):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.42)
    for i in model:
        i.fit(x_train,y_train)
        pre=i.predict(x_test)
        ac_score = accuracy_score(y_test,pre)
        print("accuracy score corresponding to random state ",r_state ,'is',ac_score)
        if ac_score>max_accuracy_score:
            max_accuracy_score=ac_score
            final_state = r_state
            final_model=i
            
print()
print()
print()
print()
print()
print()
    
print("max accuracy score corresponding to random state",final_state ,"is", max_accuracy_score,"and model is",final_model)
```

```python
#The highest value is coming on 80 random state which is 0.98 so we will take it.
```

```python
#Checking the best parameter for prediction of KNeighborsClassifier Algo using GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=80,test_size=0.42)
```

```python
KNC=KNeighborsClassifier(n_neighbors=1)
KNC.fit(x_train,y_train)
pred =KNC.predict(x_test)
acu_score = accuracy_score(y_test,pred)
print("Accuracy Score :",acu_score*100)
```

```python
score = cross_val_score(KNC,x,y,cv=5,scoring='accuracy').mean()
print('cross_val_score :',score)
print('Mean Score      :',score.mean())
print('STD score       :',score.std())
```
