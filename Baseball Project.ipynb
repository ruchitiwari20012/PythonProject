```python
x=df.iloc[:,1:]
```

```python
y=df.iloc[:,0].values
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import linear_model
```

```python
max_r_score=0
for r_state in range(42,101):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
    regr=linear_model.LinearRegression()
    regr.fit(x_train,y_train)
    y_pred = regr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print('r2_score corresponding to random state: ',r_state,'is:',r2_scr)
    if r2_scr > max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print("Max r2 score corresponding to ",final_r_state,' is ',max_r_score)
```

```python
# Lets perform the cross_validation 
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression() ,x,y,cv=5,scoring='r2').mean()
#it will calculate the 5 score and find the mean of that 
```

```python
#The below cross validation score is very less and it shows that we further need to improve our model with more data
```

```python
#since maximum r2 score is comming around random state 48 we will use that to create our final model
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=48,test_size=0.20)
lr=linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
print('r2 score is: ',r2_score(y_test,y_pred))
print("RMSE is: ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.externals import joblib
# Save the model as a pickle in a file 
joblib.dump(lr,'lr_baseball.pkl')
```
